# Задание 2.
## Составьте шаблон дизайна эксперимента для гипотезы, которая набрала больше всего баллов в практическом задании предыдущего урока

#### Шаблон гипотезы 1: 
1. Гипотеза - улучшив работу рекомендательного алгоритма мы повысим кол-во продаж;
2. Тестирование гипотезы будет проводиться на выборочной части от общего трафика;
3. Результат предположительно повлияет на следующие метрики:
    - ARPU(average revenue per user) - отношение полученного дохода к кол-ву пользователей.
4. План действий:
    Если гипотеза окажется не верной - возвращаемся к прежнему алгоритму.
    Если гипотеза окажется верной - внедряем новый алгоритм на всех пользователей.
